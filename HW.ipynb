```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
lben=preprocessing.LabelEncoder() #將文字轉換成數字 
train_data['Sex']=lben.fit_transform(train_data['Sex'])
train_data['Name']=lben.fit_transform(train_data['Name'])
train_data['Ticket']=lben.fit_transform(train_data['Ticket'])
train_data['Cabin']=lben.fit_transform(train_data['Cabin'])
train_data['Embarked']=lben.fit_transform(train_data['Embarked'])

# onehot_encoder = OneHotEncoder(sparse=False, drop='first')
# cabin_encoded = onehot_encoder.fit_transform(train_data[["Cabin"]])
# cabin_encoded_df = pd.DataFrame(cabin_encoded, columns=onehot_encoder.get_feature_names(["Cabin"]))
# train_data = pd.concat([train_data, cabin_encoded_df], axis=1)

train_data
```

```python
median_values = train_data.median()#填入中位數
train_data=train_data.fillna(median_values)
train_data
```

```python
## method 2
x=pd.DataFrame([train_data['Sex'],train_data['Age'],train_data['Ticket'],train_data['SibSp'],train_data['Pclass'],train_data['Parch'],train_data['Fare']]).T
y=train_data['Survived']
model=LogisticRegression(penalty='l2', dual=False,   
tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1,   
class_weight=None, random_state=None, solver='sag',  
max_iter=100, multi_class='auto', verbose=0,   
warm_start=False, n_jobs=None, l1_ratio=None)
model.fit(x,y)
print('訓練準確度',model.score(x,y))
```

```python
# Spliting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y, test_size=0.1, random_state = 100)

print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn import preprocessing

scaler=preprocessing.StandardScaler() #建構Normalization的物件

#把training data & test data正規化
x_train_scl=scaler.fit_transform(x_train)
x_test_scl=scaler.fit_transform(x_test)
```

```python
# other methods
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


RandomForests_Model=RandomForestClassifier()
GradientBoosting_Model=GradientBoostingClassifier(learning_rate=0.01, max_depth=100, random_state=5) #use default hyperparameters
MLP_Model=MLPClassifier(max_iter=500)
DecisionTree_Model = DecisionTreeClassifier(criterion='gini',max_depth=None)


#Training
DecisionTree_Model.fit(x_train, y_train)
RandomForests_Model.fit(x_train, y_train)
GradientBoosting_Model.fit(x_train, y_train)
MLP_Model.fit(x_train, y_train)

y_pred=DecisionTree_Model.predict(x_test)

#test
y_pred_train=DecisionTree_Model.predict(x_train)
y_pred_RF=RandomForests_Model.predict(x_test)
y_pred_GB=GradientBoosting_Model.predict(x_test)
y_pred_MLP=MLP_Model.predict(x_test)

print("Model\t\t\tTraining score\tTest score")
print("---------------------------------------------------")
print("Decision Tree:\t\t", format(accuracy_score(y_train, y_pred_train)*100,'.3f'),'%',end='') 
print("\t",format(accuracy_score(y_test, y_pred)*100,'.3f'),'%')

print("Random Forests:\t\t", format(accuracy_score(y_train, RandomForests_Model.predict(x_train))*100,'.3f'),'%',end='') 
print("\t",format(accuracy_score(y_test, y_pred_RF)*100,'.3f'),'%')

print("Gradient Boosting:\t", format(accuracy_score(y_train, GradientBoosting_Model.predict(x_train))*100,'.3f'),'%',end='') 
print("\t",format(accuracy_score(y_test, y_pred_GB)*100,'.3f'),'%')

print("Artificial NN:\t\t", format(accuracy_score(y_train, MLP_Model.predict(x_train))*100,'.3f'),'%',end='') 
print("\t",format(accuracy_score(y_test, y_pred_MLP)*100,'.3f'),'%')

```
