讀取訓練資料檔案(csv)

```python
train_data=pd.read_csv("train.csv")   # 將cvs檔案以pandas開啟
train_data
```

```python
median_values = train_data.median()#填入中位數
train_data=train_data.fillna(median_values)
train_data
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
lben=preprocessing.LabelEncoder() #將文字轉換成數字 
train_data['Sex']=lben.fit_transform(train_data['Sex'])
train_data['Name']=lben.fit_transform(train_data['Name'])
train_data['Ticket']=lben.fit_transform(train_data['Ticket'])
train_data['Cabin']=lben.fit_transform(train_data['Cabin'])
train_data['Embarked']=lben.fit_transform(train_data['Embarked'])

# onehot_encoder = OneHotEncoder(sparse=False, drop='first')
# cabin_encoded = onehot_encoder.fit_transform(train_data[["Cabin"]])
# cabin_encoded_df = pd.DataFrame(cabin_encoded, columns=onehot_encoder.get_feature_names(["Cabin"]))
# train_data = pd.concat([train_data, cabin_encoded_df], axis=1)

train_data
```

```python
x=pd.DataFrame(train_data)
y=train_data.Survived
x.head()

# Spliting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y, test_size=0.1, random_state = 100)
print(x_train.shape)
print(y_train.shape)



```

```python
model=LogisticRegression(penalty='l2', dual=False,   
tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1,   
class_weight=None, random_state=None, solver='sag',  
max_iter=100, multi_class='auto', verbose=0,   
warm_start=False, n_jobs=None, l1_ratio=None)
model.fit(x,y)
print('訓練準確度',model.score(x,y))
```

```python
test_df=pd.read_csv('test.csv')

test_df.isnull().sum()
test_df['Age']=test_df['Age'].fillna(int(train_data['Age'].median()))
test_df['Fare']=test_df['Fare'].fillna(int(train_data['Fare'].median()))
test_df['Cabin']=test_df['Cabin'].fillna(int(train_data['Cabin'].median()))

test_df['Sex'] = test_df['Sex'].astype(str)
test_df['Name'] = test_df['Name'].astype(str)
test_df['Ticket'] = test_df['Ticket'].astype(str)
test_df['Cabin'] = test_df['Cabin'].astype(str)
test_df['Embarked'] = test_df['Embarked'].astype(str)

test_df['Sex']=lben.fit_transform(test_df['Sex'])
test_df['Name']=lben.fit_transform(test_df['Name'])
test_df['Ticket']=lben.fit_transform(test_df['Ticket'])
test_df['Cabin']=lben.fit_transform(test_df['Cabin'])
test_df['Embarked']=lben.fit_transform(test_df['Embarked'])

onehot_encoder = OneHotEncoder(sparse=False, drop='first')
cabin_encoded = onehot_encoder.fit_transform(test_df[["Cabin"]])
cabin_encoded_df = pd.DataFrame(cabin_encoded, columns=onehot_encoder.get_feature_names(["Cabin"]))
test_df = pd.concat([test_df, cabin_encoded_df], axis=1)


x_test=pd.DataFrame(test_df)
y_test=model.predict(x_test)
test_df['Survived']=pd.DataFrame(y_test)
output=pd.DataFrame([test_df['PassengerId'],test_df['Survived']])
output.to_csv('sub.csv',index=False)
```
